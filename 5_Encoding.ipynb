```python
import pandas as pd
import numpy as np
import json
import re
from sklearn.preprocessing import LabelEncoder
from sklearn.cluster import KMeans

# 1. Load Data
input_path = 'ntsb_ml_filled_1.csv'
output_csv_path = 'ntsb_encoded.csv'
output_json_path = 'ntsb_encoding_map.json'

try:
    df = pd.read_csv(input_path)
    print("Data loaded successfully.")
except FileNotFoundError:
    print("File not found. Please ensure the path is correct.")
    exit()

# Dictionary to store all mapping info
encoding_maps = {}

# List to keep track of the final numeric columns we want to export
final_numeric_cols = []

# --- Helper Functions ---

def clean_aircraft_damage(val):
    if pd.isna(val): return "Unknown"
    val = str(val).lower()
    severity = {'destroyed': 4, 'substantial': 3, 'minor': 2, 'none': 1, 'unknown': 0}
    parts = [p.strip() for p in val.split(',')]
    max_score = -1
    max_label = "Unknown"
    for p in parts:
        score = severity.get(p, 0)
        if score > max_score:
            max_score = score
            max_label = p.capitalize()
    return max_label

def clean_num_engines(val):
    if pd.isna(val): return 0
    nums = re.findall(r'\d+', str(val))
    if not nums: return 0
    return sum(int(n) for n in nums)

def clean_amateur(val):
    if pd.isna(val): return "False"
    val = str(val).lower()
    if 'true' in val: return "True"
    return "False"

def clean_before_slash(val):
    if pd.isna(val): return "Unknown"
    return str(val).split('/')[0].strip()

def parse_wind(val):
    if pd.isna(val) or str(val).lower() == 'unknown':
        return 0, -1 
    val = str(val)
    knots_match = re.search(r'(\d+)\s*knots', val)
    speed = int(knots_match.group(1)) if knots_match else 0
    angle_match = re.search(r'(\d+)°', val)
    angle = int(angle_match.group(1)) if angle_match else -1
    return speed, angle

# --- Processing Columns ---

# 1. Standard Categorical Processing
simple_cats = [
    'EventType', 'City', 'Make', 'Model', 'AirCraftCategory', 
    'EngineType', 'Scheduled', 'PurposeOfFlight', 'FAR', 
    'WeatherCondition', 'Operator',  
    'Condition_of_Light', 'Visibility', 'State', 'HighestInjuryLevel',
    'Findings', 'ProbableCause', 'AirportID' # Added based on your previous prompt's context
]

for col in simple_cats:
    if col in df.columns:
        # Fill NaNs
        df[col] = df[col].astype(str).fillna('Unknown')
        
        # Encode
        le = LabelEncoder()
        encoded_col_name = f'{col}_Encoded'
        df[encoded_col_name] = le.fit_transform(df[col])
        
        # Track for output
        final_numeric_cols.append(encoded_col_name)
        
        # Save mapping
        mapping = dict(zip(le.classes_, le.transform(le.classes_)))
        encoding_maps[col] = {k: int(v) for k, v in mapping.items()}

# 2. Boolean Logic: HasSafetyRec, AmateurBuilt
df['AmateurBuilt_Clean'] = df['AmateurBuilt'].apply(clean_amateur)
df['HasSafetyRec_Clean'] = df['HasSafetyRec'].astype(str).replace({'nan': 'False'})

for col in ['AmateurBuilt_Clean', 'HasSafetyRec_Clean']:
    le = LabelEncoder()
    # Name the output column based on original name + Encoded (e.g. AmateurBuilt_Encoded)
    base_name = col.replace('_Clean', '')
    encoded_col_name = f'{base_name}_Encoded'
    
    df[encoded_col_name] = le.fit_transform(df[col])
    final_numeric_cols.append(encoded_col_name)
    
    mapping = dict(zip(le.classes_, le.transform(le.classes_)))
    encoding_maps[base_name] = {k: int(v) for k, v in mapping.items()}

# 3. Numerical Logic: NumberOfEngines
# This returns a number directly, so we use it as is
df['NumberOfEngines_Clean'] = df['NumberOfEngines'].apply(clean_num_engines)
final_numeric_cols.append('NumberOfEngines_Clean')

# 4. Complex Logic: AirCraftDamage
# Clean first (Text), then Encode (Number)
df['AirCraftDamage_Clean_Text'] = df['AirCraftDamage'].apply(clean_aircraft_damage)
le_damage = LabelEncoder()
df['AirCraftDamage_Encoded'] = le_damage.fit_transform(df['AirCraftDamage_Clean_Text'])
final_numeric_cols.append('AirCraftDamage_Encoded')

mapping = dict(zip(le_damage.classes_, le_damage.transform(le_damage.classes_)))
encoding_maps['AirCraftDamage'] = {k: int(v) for k, v in mapping.items()}

# 5. Split Logic: Lowest_Cloud_Condition, Lowest_Ceiling
for col in ['Lowest_Cloud_Condition', 'Lowest_Ceiling']:
    if col in df.columns:
        clean_text_col = df[col].apply(clean_before_slash)
        
        le = LabelEncoder()
        encoded_col_name = f'{col}_Encoded'
        df[encoded_col_name] = le.fit_transform(clean_text_col)
        final_numeric_cols.append(encoded_col_name)
        
        mapping = dict(zip(le.classes_, le.transform(le.classes_)))
        encoding_maps[col] = {k: int(v) for k, v in mapping.items()}

# 6. Wind Info (Split into Speed and Angle)
if 'Wind_Info' in df.columns:
    wind_data = df['Wind_Info'].apply(parse_wind)
    df['Wind_Speed_Knots'] = wind_data.apply(lambda x: x[0])
    df['Wind_Angle'] = wind_data.apply(lambda x: x[1])
    final_numeric_cols.extend(['Wind_Speed_Knots', 'Wind_Angle'])

# 7. Latitude / Longitude -> Region Clustering
coords = df[['Latitude', 'Longitude ']].copy() # Note the space in 'Longitude ' if present in your CSV
coords = coords.fillna(coords.mean())

kmeans = KMeans(n_clusters=5, random_state=42, n_init=10)
df['Geo_Region_Cluster'] = kmeans.fit_predict(coords)
final_numeric_cols.append('Geo_Region_Cluster')

centroids = kmeans.cluster_centers_.tolist()
encoding_maps['Geo_Region_Cluster'] = {
    "Description": "KMeans Clustering on Lat/Lon (k=5)",
    "Centroids_Lat_Lon": centroids
}

# 8. Handling Injury Counts
injury_cols = ['FatalInjuryCount', 'SeriousInjuryCount', 'MinorInjuryCount', 'OnboardInjuryCount', 'OnGroundInjuryCount']
for col in injury_cols:
    if col in df.columns:
        df[col] = df[col].fillna(0).astype(int)
        final_numeric_cols.append(col)

# --- Final Export ---

# Create final dataframe with NtsbNo and ONLY the tracked numeric columns
# We use a set to remove potential duplicates if logic ran twice, then convert to list
cols_to_export = ['NtsbNo'] + list(set(final_numeric_cols))

# Check if NtsbNo exists (it should), if not, skip it
if 'NtsbNo' not in df.columns:
    cols_to_export.remove('NtsbNo')

df_final = df[cols_to_export].copy()

df_final.to_csv(output_csv_path, index=False)
print(f"Encoded CSV saved to: {output_csv_path}")
print(f"Total columns in output: {len(df_final.columns)}")
print("Sample of columns:", df_final.columns.tolist()[:5])

# Save Mapping JSON
with open(output_json_path, 'w') as f:
    json.dump(encoding_maps, f, indent=4)
print(f"Mapping JSON saved to: {output_json_path}")
```
